# Apply filter in CNN model

### A filter is a small matrix of weights that slides over the input image (or feature map) to extract patterns.

#### ➤ Key points:
* It has a size (e.g., 3×3, 5×5, etc.).
* It detects features like edges, textures, or patterns.
* Each filter produces one feature map (activation map).
* The number of filters = number of output channels after the convolution layer.

### here we apply filter in CNN model
* no.of convoluation layer=3
* no.of filter=32
* filter size=(3,3)
* padding=None

```python
model_filter=Sequential()
model_filter.add(Conv2D(32,kernel_size=(3,3),padding='valid',activation='relu',input_shape=(28,28,1)))
model_filter.add(Conv2D(32,kernel_size=(3,3),padding='valid',activation='relu'))
model_filter.add(Conv2D(32,kernel_size=(3,3),padding='valid',activation='relu'))

model_filter.add(Flatten())
model_filter.add(Dense(128,activation='relu'))
model_filter.add(Dense(10,activation='softmax'))
```

```python
model_filter.summary()
```

### model_filter summary explaination
 * after appling first convolution layer(Conv2d_1) image shape(28,28,) is reduced to (26,26,32)
 * because filter formula is :(**image_size-filter_size+1**) i.e(28-3+1)=(26,26,32)
 * here in (26,26,32) `32` is the no. of filter used
 * in same way it is reduced in layer `2nd` and `3rd`

# Apply padding in CNN model

### Padding means adding zeros (or other values) around the border of the input image before applying convolution.

#### ➤ Why use padding:

* To preserve spatial size after convolution.
* To avoid losing edge information (because filters “fall off” the edges).
* To control the output dimensions.

```python
model_padding=Sequential()
model_padding.add(Conv2D(32,kernel_size=(3,3),padding='same',activation='relu',input_shape=(28,28,1)))
model_padding.add(Conv2D(32,kernel_size=(3,3),padding='same',activation='relu'))
model_padding.add(Conv2D(32,kernel_size=(3,3),padding='same',activation='relu'))

model_padding.add(Flatten())
model_padding.add(Dense(128,activation='relu'))
model_padding.add(Dense(10,activation='softmax'))
```

```python
model_padding.summary()
```

### model_padding summary explanation
* here we can see the output shape of all three convolution layers is same because of padding
* it is calculated by formula:(**image_shape+2*padding_size-filter_size+1**)
* i.e (28-2*1-3+1)=28
* padding_size is 1

# Apply Stride in CNN model

### Stride refers to how many pixels the convolution filter (kernel) moves across the input image (or feature map) at each step.

#### Advantages of Higher Stride
##### 1.Reduces computation
* Fewer filter operations → faster training and inference.
##### 2.Reduces memory usage
* Smaller output feature maps.
##### 3. Acts like downsampling (similar to pooling)
* Reduces spatial dimensions early in the network.
##### 4.Helps avoid overfitting (sometimes)
* Because it reduces model complexity.

#### Disadvantages of Higher Stride

##### 1.Loss of spatial information
* Skipping pixels means the network might miss fine details.
##### 2.Less precise feature localization
* Useful for classification but bad for tasks like segmentation or object detection.
##### 3.May cause underfitting
* The model might not capture small or subtle patterns.

```python
model_stride=Sequential()
model_stride.add(Conv2D(32,kernel_size=(3,3),padding='same',strides=(2,2),activation='relu',input_shape=(28,28,1)))
model_stride.add(Conv2D(32,kernel_size=(3,3),padding='same',strides=(2,2),activation='relu'))
model_stride.add(Conv2D(32,kernel_size=(3,3),padding='same',strides=(2,2),activation='relu'))

model_stride.add(Flatten())
model_stride.add(Dense(128,activation='relu'))
model_stride.add(Dense(10,activation='softmax'))
```

```python
model_stride.summary()
```

### model_stride summary explanation
* formula:**((Input_size+2*padding_size-filter_size)/stride_size)+1)**
* math.floor((28-2*1-3)/2+1)=14 for fist convolution layer
* input_size=28
* padding_size=(1,1)
* filter_size=(3,3)
* stride_size=(2,2)

# Apply pooling in CNN model

#### **pooling** (also called subsampling or downsampling) is an operation used to reduce the spatial dimensions (width and height) of the feature maps while retaining the most important information.
##### Purpose of Pooling
###### 1.Dimensionality Reduction	
* Reduces the size of feature maps, leading to fewer parameters and less computation.
##### 2.Translation Invariance	
* Small movements or distortions in the image have little effect on pooled features.
###### 3. Prevent Overfitting	
* By simplifying the network and reducing the number of parameters.

```python
model_pooling=Sequential()
model_pooling.add(Conv2D(32,kernel_size=(3,3),padding='valid',activation='relu',input_shape=(28,28,1)))
model_pooling.add(MaxPooling2D(pool_size=(2,2),strides=2,padding='valid'))
model_pooling.add(Conv2D(32,kernel_size=(3,3),padding='valid',activation='relu'))
model_pooling.add(MaxPooling2D(pool_size=(2,2),strides=2,padding='valid'))

model_pooling.add(Flatten())
model_pooling.add(Dense(128,activation='relu'))
model_pooling.add(Dense(10,activation='softmax'))
                  
```

```python
model_pooling.summary()
```

#### model_pooling summary explanation
* in first layer i.e `Conv2d_30` output shape is (None,26,26,32)
* because filter_size(kernel_size) is (3,3)
* in second layer i.e `MaxPooling2D` output shape is (None,13,13,32)
* because (2,2) maxpooling is applied i.e 26/2
